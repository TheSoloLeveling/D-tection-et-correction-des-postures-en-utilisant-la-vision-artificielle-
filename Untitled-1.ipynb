```python
import mediapipe as mp
import cv2
```

```python
#set up drawing utilities and load holistic model
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic
```

```python
cap = cv2.VideoCapture('http://192.168.11.104:4747/video')

#set up holistic model, parameters ==> tracking accuracy 
with mp_holistic.Holistic(min_detection_confidence=0.5, min_tracking_confidence=0.5) as holistic:

    while cap.isOpened():
        ret, frame = cap.read()

        #holistic takes as input RGB image
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        results = holistic.process(image)

        #make detections
        print(results.left_hand_landmarks)

        #convert to BGR for rendering
        image = cv2.cvtColor(frame, cv2.COLOR_RGB2BGR)

        #draw face landmarks
        #mp_holistic.FACE_CONNECTIONS ==> shows the relationship between all of the different components ,connected points (mesh)
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS)

        cv2.imshow('Holistic Model Detections', image)
        if cv2.waitKey(10) & 0xFF == ord('q'):
            break

cap.release()
cv2.destroyAllWindows()
```
